### Win percentage of each team

```python
match=(df['team1'].value_counts()+df['team2'].value_counts()).reset_index()
match.columns=['teams','matches']
match
```

```python
df['winner']=df['winner'].map(Teams)
```

```python
wins=df['winner'].value_counts().reset_index()
wins.columns=['teams','wins']
wins
```

```python
df_matches=match.merge(wins,on='teams',how='inner')
df_matches
```

```python
df_matches['%win']=(df_matches['wins']/df_matches['matches'])*100
```

```python
df_matches
```

```python
plt.figure(figsize=(8,6))
plt.bar(df_matches['teams'],df_matches['matches'],label='Played')
plt.bar(df_matches['teams'],df_matches['wins'],label='Won')
plt.xlabel('Teams')
plt.ylabel('Total matches')
plt.title('Stacked plot of matches played and matches won')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.bar(df_matches['teams'],df_matches['%win'])
plt.xlabel('Teams')
plt.ylabel('Win Percentage')
plt.title('Win percentage of each team')
plt.show()
```

### Total runs in each season

```python
df.head()
```

```python
df2=pd.read_csv("deliveries.csv")
df2.head()
```

```python
season=df[['id','season']].merge(df2,left_on='id',right_on='match_id',how='left').drop('id',axis=1)
season
```

```python
season=season.groupby('season')['total_runs'].sum().reset_index()
season
```

```python
plt.plot(season['season'],season['total_runs'])
```

```python
avg_runs=df.groupby('season')['id'].count().reset_index().rename(columns={'id':'matches'})
avg_runs
```

```python
season=pd.concat([avg_runs,season.iloc[:,1]],axis=1)
season
```

```python
season['per_match_runs']=season['total_runs']/season['matches']
```

```python
season
```

```python
## Average runs in each match in each season
plt.plot(season['season'],season['per_match_runs'])
```

```python
## Lucky venue for each team
def lucky(df,team_name):
    return df[df['winner']==team_name]['venue'].value_counts().nlargest(5)
```

```python
lucky(df,'CSK').plot.bar()
```

```python
## Comparison between two teams
def comparison(team1,team2):
    compare=df[((df['team1']==team1)|(df['team2']==team1))&((df['team1']==team2)|(df['team2']==team2))]
    sns.countplot(x='season',hue='winner',data=compare)
```

```python
comparison('RCB','CSK')
```
